## Importemos las librerias requeridas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Procedemos a la lectura de las hojas del archivo Excel

```python
df2 = pd.read_excel("homicidios.xlsx",sheet_name=2)
df2
```

### **1. Empezemos observando la data del primer dataframe**

```python
df.describe(include='all')
```

```python
df.columns
```

#### Observamos que la columna donde se encuentra la data relevante es la cantidad de columnas, observamos que en la columna **N_VICTIMAS** el promedio del numero de victimas es 1, ademas observamos una mediana aproximada de 7 en la columna **COMUNA**

```python
df.isnull().sum()
```

#### La columna **Altura** es la que presenta mas del **80%** de los datos faltantes

#### Habiamos concluido que en la columna **COMUNA** existia la mediana cercana a 7, observemos la distribución de los datos

```python
plt.hist(df["COMUNA"])
plt.show()
```

#### Para las columnas categoricas como **TIPO_DE_CALLE, VICTIMA, ACUSADO** veremos la distribución de las mismas, pues esto ayuda a observar mejor los datos

```python
plt.hist(df["TIPO_DE_CALLE"])
plt.show()
```

#### Para esta columna observamos que la mayoria de accidentes suceden en *AVENIDAS*

```python
plt.hist(df["VICTIMA"],bins=10)
plt.xticks(rotation=45)
plt.show()
```

#### Para esta columna observamos que en la mayoria de accidentes los afectados son **motos(conductores de motos o sus ocupantes) y peatones**

```python
plt.hist(df["ACUSADO"],bins=10)
plt.xticks(rotation=45)
plt.show()
```

#### En este caso observamos que los 3 principales acusados de los accidentes son: **AUTO, PASAJEROS(DE UN AUTOBUS PÚBLICO) Y CARGAS**

## *Ahora observemos los datos presentes en el segundo dataframe*

```python
df2.describe(include="all")
```

```python
df2["ID_hecho"].describe()
```

#### Observamos 3 variables categoricas, en las cuales podriamos sacar el valor que mas se repite, estas columnas son: **ROL, VICTIMA y  SEXO**

```python
plt.hist(df2["ROL"],bins=15)
plt.xticks(rotation=45)
plt.show()
```

#### Esta grafica nos muestra que las personas que eran *CONDUCTORES* o *PEATONES* son quienes presentan mayor recurrencia en la data de fallecidos

```python
plt.hist(df2["VICTIMA"],bins=10)
plt.xticks(rotation=45)
plt.show()
```

#### Al igual que el primer dataframe observamos que la mayoria de ocurrencias sucede con *MOTOS* y *PEATONES*

```python
plt.hist(df2["SEXO"],bins=5)
plt.xlabel("SEXO")
plt.show()
```

#### Finalmente observamos que segun registro, las personas de sexo **MASCULINO** son las que presentan mayor cantidad de decesos
